# Drug Discovery-AI : Nettoyage et Prétraitement de Données de Bio-informatique

Ce notebook présente un pipeline complet, structuré et professionnel pour le nettoyage, la validation moléculaire et le prétraitement de deux jeux de données de bio-informatique :
1. **BACE** (`bace.csv`) : Un jeu de données de classification moléculaire (activité biologique binaire contre la cible BACE-1).
2. **DAVIS** (`davis.csv`) : Un jeu de données de régression (affinité de liaison protéine-ligand, mesurée par la constante de dissociation $K_d$ en nM).

### Objectifs
- Valider et canonicaliser les structures chimiques (SMILES) avec **RDKit**.
- Générer de nouveaux descripteurs moléculaires physico-chimiques fondamentaux.
- Réaliser une analyse exploratoire et statistique robuste des données.
- Traiter les valeurs aberrantes (outliers) et manquantes de façon argumentée.
- Encoder les variables catégorielles et standardiser les descripteurs.
- Gérer le déséquilibre de classes via la technique de suréchantillonnage **SMOTE**.
- Effectuer une réduction de dimensionnalité exploratoire (**PCA 2D**).
- Séparer et sauvegarder les jeux de données nettoyés et prêts pour la modélisation.

*Ce pipeline adopte le style académique rigoureux exigé dans un projet de niveau Licence en Intelligence Artificielle.*

2. Les doublons moléculaires (les molécules ayant le même SMILES).

Par conséquent, **aucun doublon ne doit être supprimé** dans DAVIS, car cela détruirait des couples uniques d'interaction drogue-protéine.


## 5. Validation des SMILES avec RDKit

Avant tout calcul chémo-informatique, nous devons nous assurer que chaque chaîne SMILES correspond à une molécule chimiquement valide. Nous utilisons la fonction `Chem.MolFromSmiles()` de **RDKit**. Si cette fonction renvoie `None` (ou échoue à la validation de la structure tridimensionnelle/valence), la molécule est jugée invalide et exclue.


```python
def valider_smiles(df, smiles_col):
    """
    Parcourt le dataframe, valide chaque chaîne SMILES avec RDKit,
    et filtre les lignes contenant des molécules invalises.
    """
    taille_initiale = len(df)
    
    # Créer un masque booléen de validité
    est_valide = df[smiles_col].apply(lambda s: Chem.MolFromSmiles(s) is not None)
    
    # Filtrer le dataframe
    df_nettoye = df[est_valide].copy()
    nb_supprimes = taille_initiale - len(df_nettoye)
    
    return df_nettoye, nb_supprimes

print("Validation du dataset BACE...")
df_bace, bace_invalides = valider_smiles(df_bace, 'mol')
print(f"BACE : {bace_invalides} molécules invalides ont été supprimées.")

print("\nValidation du dataset DAVIS...")
df_davis, davis_invalides = valider_smiles(df_davis, 'Drug')
print(f"DAVIS : {davis_invalides} molécules invalides ont été supprimées.")
```

## 6. Canonicalisation des SMILES

Une structure chimique peut posséder plusieurs chaînes SMILES distinctes (différences dans l'ordre de parcours des atomes). La **canonicalisation** convertit chaque SMILES vers une représentation normalisée unique (le SMILES canonique) grâce à l'algorithme standardisé implémenté dans `Chem.MolToSmiles()`. Cela uniformise les chaînes et prévient les représentations multiples redondantes.


```python
def canonicaliser_smiles(df, smiles_col):
    """Convertit toutes les chaînes SMILES d'un DataFrame en SMILES canoniques."""
    def to_canonical(smiles_str):
        mol = Chem.MolFromSmiles(smiles_str)
        if mol is not None:
            return Chem.MolToSmiles(mol, canonical=True)
        return None
    
    df_res = df.copy()
    df_res[smiles_col] = df_res[smiles_col].apply(to_canonical)
    return df_res

print("Canonicalisation des molécules de BACE...")
df_bace = canonicaliser_smiles(df_bace, 'mol')

print("Canonicalisation des molécules de DAVIS...")
df_davis = canonicaliser_smiles(df_davis, 'Drug')

print("\nExemple de SMILES canonique BACE :")
print(df_bace['mol'].iloc[0])
print("\nExemple de SMILES canonique DAVIS :")
print(df_davis['Drug'].iloc[0])
```

## 7. Création de nouvelles features moléculaires avec RDKit

Les chaînes de caractères SMILES ne peuvent pas servir directement d'entrées pour les modèles de Machine Learning traditionnels. Nous générons 8 descripteurs physico-chimiques fondamentaux, inspirés des règles de Lipinski, pour représenter les molécules :
1. **Molecular Weight (MW)** : Masse moléculaire (taille de la molécule).
2. **LogP (MolLogP)** : Lipophilie estimée.
3. **TPSA** : Surface polaire topologique (indicateur d'absorption intestinale et de pénétration de la barrière hémato-encéphalique).
4. **NumHDonors** : Nombre de donneurs de liaisons hydrogène.
5. **NumHAcceptors** : Nombre d'accepteurs de liaisons hydrogène.
6. **Rotatable Bonds** : Nombre de liaisons rotatives (flexibilité moléculaire).
7. **Ring Count** : Nombre de cycles aromatiques ou aliphatiques.
8. **Heavy Atom Count** : Nombre d'atomes lourds (non-hydrogène).


```python
# Définir le nom des features
feature_cols = ['MW', 'LogP', 'TPSA', 'NumHDonors', 'NumHAcceptors', 'RotatableBonds', 'RingCount', 'HeavyAtomCount']

def extraire_descripteurs(df, smiles_col):
    """
    Calcule les 8 descripteurs moléculaires physico-chimiques
    pour chaque molécule unique du DataFrame.
    """
    # Extraire les SMILES uniques pour optimiser le temps de calcul
    smiles_uniques = df[smiles_col].unique()
    
    dict_features = {}
    for smiles in smiles_uniques:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            dict_features[smiles] = {
                'MW': Descriptors.MolWt(mol),
                'LogP': Descriptors.MolLogP(mol),
                'TPSA': Descriptors.TPSA(mol),
                'NumHDonors': Descriptors.NumHDonors(mol),
                'NumHAcceptors': Descriptors.NumHAcceptors(mol),
                'RotatableBonds': Descriptors.NumRotatableBonds(mol),
                'RingCount': Descriptors.RingCount(mol),
                'HeavyAtomCount': Descriptors.HeavyAtomCount(mol)
            }
            
    # Convertir en DataFrame
    df_features = pd.DataFrame.from_dict(dict_features, orient='index').reset_index()
    df_features.rename(columns={'index': smiles_col}, inplace=True)
    return df_features

print("Calcul des descripteurs moléculaires pour BACE...")
features_bace = extraire_descripteurs(df_bace, 'mol')

# Éliminer les colonnes existantes dans df_bace portant le même nom pour éviter les suffixes (_x, _y) lors de la fusion
cols_overlap_bace = [c for c in feature_cols if c in df_bace.columns]
if cols_overlap_bace:
    df_bace = df_bace.drop(columns=cols_overlap_bace)

df_bace = df_bace.merge(features_bace, on='mol', how='left')

print("Calcul des descripteurs moléculaires pour DAVIS...")
features_davis = extraire_descripteurs(df_davis, 'Drug')

# Éliminer les colonnes existantes dans df_davis portant le même nom pour éviter les collisions
cols_overlap_davis = [c for c in feature_cols if c in df_davis.columns]
if cols_overlap_davis:
    df_davis = df_davis.drop(columns=cols_overlap_davis)

df_davis = df_davis.merge(features_davis, on='Drug', how='left')

print("\n=== Aperçu des caractéristiques moléculaires dans BACE ===")
display(df_bace[['mol'] + feature_cols].head(3))
```

## 9. Traitement des outliers

Les boxplots révèlent la présence de valeurs atypiques dans plusieurs descripteurs (ex. des poids moléculaires ou des surfaces polaires TPSA particulièrement élevés).
Nous comparons deux approches classiques :
1. **Écart Interquartile (IQR)** : Supprime les lignes dont les valeurs sont en dehors de l'intervalle $[Q_1 - 1.5 \times IQR, Q_3 + 1.5 \times IQR]$.
2. **Winsorisation (Clipping)** : Remplace les valeurs extrêmes situées au-delà de centiles de sécurité (ici le 1er centile pour le bas et le 99e centile pour le haut) par la valeur de ces centiles.

### Choix de la méthode et Justification
En bio-informatique, nous sélectionnons la **Winsorisation (clipping)**.
- **Justification** : La suppression de lignes (via l'IQR) éliminerait des composés chimiques réels, volumineux ou hautement hydrophobes, qui sont des candidats médicaments légitimes (comme des structures peptidomimétiques ou macrocycliques).
- Supprimer ces molécules induirait une perte d'information biologique critique.
- La Winsorisation permet de neutraliser l'impact disproportionné des valeurs extrêmes sur les modèles (comme les modèles linéaires ou de régression) en "bornant" les extrêmes sans sacrifier aucune ligne de données.


```python
def appliquer_winsorisation(df, cols, bas=0.01, haut=0.99):
    """Applique le clipping aux centiles [bas, haut] pour les colonnes choisies."""
    df_treated = df.copy()
    for col in cols:
        q_bas = df[col].quantile(bas)
        q_haut = df[col].quantile(haut)
        df_treated[col] = df_treated[col].clip(lower=q_bas, upper=q_haut)
    return df_treated

print("Application de la Winsorisation sur BACE (centiles 1% et 99%)...")
df_bace_treated = appliquer_winsorisation(df_bace, feature_cols)

print("Application de la Winsorisation sur DAVIS (centiles 1% et 99%)...")
df_davis_treated = appliquer_winsorisation(df_davis, feature_cols)

print("✓ Traitement des outliers terminé par Winsorisation.")
```

## 10. Encodage des colonnes catégorielles

Pour que les données catégorielles (non numériques) soient compréhensibles par les modèles d'apprentissage, nous devons les encoder en nombres.
- **BACE** : La colonne `Model` contient des valeurs indiquant le partitionnement initial (`Train`, `Valid`, `Test`). Bien qu'il ne s'agisse pas d'une feature prédictive mais d'un indicateur de découpage, nous démontrons son encodage via le **One-Hot Encoding** (création de colonnes binaires).
- **DAVIS** : La colonne `Target_ID` indique le nom de la protéine cible (379 uniques). Faire du One-Hot Encoding créerait 379 colonnes éparses, ce qui est sous-optimal. Nous appliquons le **Label Encoding** pour assigner un entier unique compris entre 0 et 378 à chaque protéine.


```python
# One-Hot Encoding pour 'Model' dans BACE
print("=== Encodage BACE (One-Hot) ===")
df_bace_encoded = pd.get_dummies(df_bace_treated, columns=['Model'], prefix='Model', drop_first=False)
# Convertir les colonnes booléennes résultantes en entiers (0/1)
one_hot_cols = [c for c in df_bace_encoded.columns if 'Model_' in c]
df_bace_encoded[one_hot_cols] = df_bace_encoded[one_hot_cols].astype(int)
print(f"Colonnes générées pour BACE : {one_hot_cols}")

# Label Encoding pour 'Target_ID' dans DAVIS
print("\n=== Encodage DAVIS (Label Encoding) ===")
le_target = LabelEncoder()
df_davis_encoded = df_davis_treated.copy()
df_davis_encoded['Target_ID_encoded'] = le_target.fit_transform(df_davis_encoded['Target_ID'])
print("Correspondance des 3 premières protéines encodées :")
display(df_davis_encoded[['Target_ID', 'Target_ID_encoded']].drop_duplicates().head(3))
```

## 11. Normalisation et Standardisation

Les 8 descripteurs moléculaires physico-chimiques calculés possèdent des plages de valeurs très différentes (ex. la masse moléculaire MW varie de 100 à 1000, tandis que le nombre de liaisons donneuses H-Donors varie de 0 à 10). Nous comparons deux techniques courantes :
1. **StandardScaler** : Centre et réduit les variables (moyenne = 0, écart-type = 1).
2. **MinMaxScaler** : Met à l'échelle dans l'intervalle $[0, 1]$.

### Comparaison visuelle et choix de la méthode
Nous testons les deux méthodes sur le dataset BACE et visualisons l'effet sur le poids moléculaire (MW) et la lipophilie (LogP).
Nous choisirons la **Standardisation (StandardScaler)** pour le traitement final.
- **Justification** : Contrairement au MinMaxScaler qui compresse artificiellement les distributions dans des limites strictes (et dont la formule dépend fortement des valeurs maximales/minimales), le StandardScaler conserve les proportions relatives de variance et ne modifie pas la forme de la distribution (la conservant centrée sur sa moyenne), ce qui est optimal pour la plupart des algorithmes d'apprentissage (SVM, régression logistique, réseaux de neurones).


```python
# Extraire les features brutes à normaliser
X_bace_raw = df_bace_encoded[feature_cols]

# Initialiser les scalers
scaler_std = StandardScaler()
scaler_minmax = MinMaxScaler()

X_bace_std = scaler_std.fit_transform(X_bace_raw)
X_bace_minmax = scaler_minmax.fit_transform(X_bace_raw)

# Visualisation comparative
fig, axes = plt.subplots(2, 3, figsize=(16, 9))

# MW
sns.histplot(X_bace_raw['MW'], kde=True, color="gray", ax=axes[0, 0])
axes[0, 0].set_title("MW - Original", fontsize=11, fontweight='bold')
sns.histplot(X_bace_std[:, 0], kde=True, color="#4A90E2", ax=axes[0, 1])
axes[0, 1].set_title("MW - StandardScaler", fontsize=11, fontweight='bold')
sns.histplot(X_bace_minmax[:, 0], kde=True, color="#50E3C2", ax=axes[0, 2])
axes[0, 2].set_title("MW - MinMaxScaler", fontsize=11, fontweight='bold')

# LogP
sns.histplot(X_bace_raw['LogP'], kde=True, color="gray", ax=axes[1, 0])
axes[1, 0].set_title("LogP - Original", fontsize=11, fontweight='bold')
sns.histplot(X_bace_std[:, 1], kde=True, color="#4A90E2", ax=axes[1, 1])
axes[1, 1].set_title("LogP - StandardScaler", fontsize=11, fontweight='bold')
sns.histplot(X_bace_minmax[:, 1], kde=True, color="#50E3C2", ax=axes[1, 2])
axes[1, 2].set_title("LogP - MinMaxScaler", fontsize=11, fontweight='bold')

plt.suptitle("Impact des méthodes d'échelle (StandardScaler vs MinMaxScaler)", y=1.01, fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()

# Application finale du StandardScaler sur les deux jeux de données
print("Normalisation avec StandardScaler en cours...")
df_bace_scaled = df_bace_encoded.copy()
df_bace_scaled[feature_cols] = StandardScaler().fit_transform(df_bace_encoded[feature_cols])

df_davis_scaled = df_davis_encoded.copy()
df_davis_scaled[feature_cols] = StandardScaler().fit_transform(df_davis_encoded[feature_cols])
print("✓ Les descripteurs de BACE et DAVIS ont été mis à l'échelle via StandardScaler.")
```

## 12. Gestion du déséquilibre des classes (pour BACE)

Un déséquilibre prononcé des classes peut perturber l'apprentissage et biaiser les métriques de performance.
- Nous calculons d'abord le ratio de la classe active (1) sur la classe inactive (0) dans BACE.
- Si le déséquilibre est avéré, nous configurons l'application de la méthode **SMOTE** (Synthetic Minority Over-sampling Technique) qui génère des molécules virtuelles synthétiques dans le voisinage de la classe minoritaire.

> [!IMPORTANT]
> **Règle méthodologique cruciale :** Afin d'éviter tout phénomène de fuite de données (*data leakage*), la méthode SMOTE ne doit **jamais** être appliquée sur la totalité du jeu de données ou sur le jeu de test. Elle doit être appliquée **uniquement sur le jeu d'entraînement** issu de la séparation Train/Test. Dans cette section, nous démontrons le fonctionnement théorique de SMOTE, puis nous l'appliquerons rigoureusement après la séparation dans la section 14.


```python
# Analyse de la distribution de Class dans BACE
comptage_classes = df_bace_scaled['Class'].value_counts()
ratio_classes = comptage_classes[1] / comptage_classes[0]

print(f"Distribution initiale des classes (BACE) :")
print(f"  - Inactifs (Classe 0) : {comptage_classes[0]} ({comptage_classes[0]/len(df_bace_scaled)*100:.1f}%)")
print(f"  - Actifs (Classe 1)   : {comptage_classes[1]} ({comptage_classes[1]/len(df_bace_scaled)*100:.1f}%)")
print(f"  - Ratio Actifs / Inactifs : {ratio_classes:.3f}")

# Démonstration globale de SMOTE
smote_global = SMOTE(random_state=42)
X_bace_demo, y_bace_demo = smote_global.fit_resample(df_bace_scaled[feature_cols], df_bace_scaled['Class'])

print("\n=== Démonstration de SMOTE ===")
print(f"Dimensions de X avant SMOTE : {df_bace_scaled[feature_cols].shape}")
print(f"Dimensions de X après SMOTE : {X_bace_demo.shape}")
print("Distribution des classes après SMOTE :")
print(y_bace_demo.value_counts())
```

## 13. Réduction dimensionnelle exploratoire (PCA 2D)

Pour appréhender la structure globale des molécules et vérifier si nos descripteurs permettent de différencier le comportement biologique, nous réalisons une **Analyse en Composantes Principales (PCA)** à 2 dimensions sur nos 8 descripteurs physico-chimiques standardisés.


```python
# 13.1 PCA sur BACE
pca_bace = PCA(n_components=2, random_state=42)
bace_pca_proj = pca_bace.fit_transform(df_bace_scaled[feature_cols])

df_bace_pca = pd.DataFrame(bace_pca_proj, columns=['PC1', 'PC2'])
df_bace_pca['Class'] = df_bace_scaled['Class'].values

# 13.2 PCA sur DAVIS
pca_davis = PCA(n_components=2, random_state=42)
davis_pca_proj = pca_davis.fit_transform(df_davis_scaled[feature_cols])

df_davis_pca = pd.DataFrame(davis_pca_proj, columns=['PC1', 'PC2'])
df_davis_pca['Affinite_Y'] = df_davis_scaled['Y'].values
# On utilise le log10 de Y pour compresser la plage dynamique pour la couleur
df_davis_pca['log10_Y'] = np.log10(df_davis_pca['Affinite_Y'] + 1e-5)

# Visualisation des projections PCA 2D
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# BACE PCA Plot
sns.scatterplot(x='PC1', y='PC2', hue='Class', data=df_bace_pca, palette="Set1", alpha=0.7, ax=axes[0])
var_bace_1 = pca_bace.explained_variance_ratio_[0]*100
var_bace_2 = pca_bace.explained_variance_ratio_[1]*100
axes[0].set_title(f"PCA 2D - BACE (Variance expliquée cumulative: {var_bace_1+var_bace_2:.1f}%)", fontsize=13, fontweight='bold')
axes[0].set_xlabel(f"PC1 ({var_bace_1:.1f}%)")
axes[0].set_ylabel(f"PC2 ({var_bace_2:.1f}%)")

# DAVIS PCA Plot
scatter = axes[1].scatter(df_davis_pca['PC1'], df_davis_pca['PC2'], c=df_davis_pca['log10_Y'], cmap="viridis", alpha=0.5, s=15)
fig.colorbar(scatter, ax=axes[1], label="Affinité log10(Y)")
var_davis_1 = pca_davis.explained_variance_ratio_[0]*100
var_davis_2 = pca_davis.explained_variance_ratio_[1]*100
axes[1].set_title(f"PCA 2D - DAVIS (Variance expliquée cumulative: {var_davis_1+var_davis_2:.1f}%)", fontsize=13, fontweight='bold')
axes[1].set_xlabel(f"PC1 ({var_davis_1:.1f}%)")
axes[1].set_ylabel(f"PC2 ({var_davis_2:.1f}%)")

plt.tight_layout()
plt.show()
```

## 14. Découpage Train / Test

Nous découpons les deux jeux de données afin de pouvoir évaluer ultérieurement nos modèles prédictifs.
- **BACE (Classification)** : Nous utilisons une séparation **stratifiée** (`Stratified Split` avec l'argument `stratify=y`) afin que les proportions de molécules actives (1) et inactives (0) soient identiques dans les jeux de Train et de Test. C'est sur ce jeu de Train ainsi isolé que nous appliquons **SMOTE** pour équilibrer parfaitement les classes.
- **DAVIS (Régression)** : Nous effectuons une séparation classique aléatoire simple des interactions.


```python
# 14.1 Stratified Split & SMOTE pour BACE
X_bace = df_bace_scaled[feature_cols + one_hot_cols]
y_bace = df_bace_scaled['Class']

X_bace_train_raw, X_bace_test, y_bace_train_raw, y_bace_test = train_test_split(
    X_bace, y_bace, test_size=0.2, random_state=42, stratify=y_bace
)

# Application de SMOTE sur le jeu d'entraînement uniquement (Rigueur académique)
smote_obj = SMOTE(random_state=42)
X_bace_train, y_bace_train = smote_obj.fit_resample(X_bace_train_raw, y_bace_train_raw)

print("=== DECOUPAGE ET EQUILIBRAGE - BACE ===")
print(f"Jeu d'entraînement original : {X_bace_train_raw.shape[0]} molécules")
print(f"Jeu d'entraînement après SMOTE : {X_bace_train.shape[0]} molécules")
print(f"  - Inactifs (Classe 0) : {sum(y_bace_train == 0)}")
print(f"  - Actifs (Classe 1) : {sum(y_bace_train == 1)}")
print(f"Jeu de test (stratifié, non bruité par SMOTE) : {X_bace_test.shape[0]} molécules")
print(f"  - Distribution des classes dans Test (proportion) :\n{y_bace_test.value_counts(normalize=True)}")

# 14.2 Split Classique pour DAVIS
# Les features incluent les descripteurs standardisés de la molécule plus le Target_ID encodé
features_davis_cols = feature_cols + ['Target_ID_encoded']
X_davis = df_davis_scaled[features_davis_cols]
y_davis = df_davis_scaled['Y']

X_davis_train, X_davis_test, y_davis_train, y_davis_test = train_test_split(
    X_davis, y_davis, test_size=0.2, random_state=42
)

print("\n=== DECOUPAGE - DAVIS ===")
print(f"Jeu d'entraînement : {X_davis_train.shape[0]} couples drogue-cible")
print(f"Jeu de test : {X_davis_test.shape[0]} couples drogue-cible")
```
